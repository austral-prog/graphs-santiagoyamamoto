```python
import matplotlib.pyplot as plt
import numpy as np
import os

if not os.path.exists('imgs'):
    os.makedirs('imgs')
```

### 1. Gráfico de Líneas: Tiempos de Vuelta en Fórmula 1
Este gráfico de líneas muestra la evolución del tiempo en segundos durante 5 vueltas consecutivas para dos pilotos diferentes, Franco Colapinto y Pierre Gasly. Elegí el gráfico de líneas porque permite ver facilmente como evoluciona una variable con respecto a otra, en este caso el tiempo de vuelta con el número de vuelta.

```python
vueltas = [1, 2, 3, 4, 5]
tiempos_colapinto = [85.2, 84.8, 84.5, 84.1, 83.9]
tiempos_gasly = [85.5, 85.0, 84.7, 84.6, 84.2]

plt.figure(figsize=(8, 5))

plt.plot(vueltas, tiempos_colapinto, color='blue', linestyle='-', marker='o', label='Colapinto')
plt.plot(vueltas, tiempos_gasly, color='red', linestyle='--', marker='x', label='Gasly')

plt.title('Evolución de Tiempos de Vuelta')
plt.xlabel('Número de Vuelta')
plt.ylabel('Tiempo (segundos)')
plt.legend()
plt.grid(True, linestyle=':') #

plt.savefig('imgs/colapinto', dpi=300, bbox_inches='tight')
plt.show()
```

### 2. Gráfico de Barras: Goles a favor en el torneo
Este gráfico de barras compara la cantidad total de goles anotados por cuatro equipos de fútbol diferentes, destacando a Boca Jrs, en los últimos 10 partidos. Elegí el gráfico de barras porque permite comparar rápidamente, en este caso, quien hizo más y menos goles.

```python
equipos = ['Boca Juniors', 'Independiente', 'Racing', 'Riber']
goles = [18, 14, 12, 1]
colores = ['#004488', 'red', 'lightblue', 'white'] 

plt.figure(figsize=(8, 5))

plt.bar(equipos, goles, color=colores, edgecolor='black', label='Goles Totales')

plt.title('Goles a favor en los últimos 10 partidos')
plt.xlabel('Equipos')
plt.ylabel('Cantidad de Goles')
plt.legend()

plt.savefig('imgs/boca.png', dpi=300, bbox_inches='tight')
plt.show()
```

### 3. Gráfico de Dispersión: Precisión vs Daño en el juego Valorant
Este gráfico muestra la relación entre el porcentaje de tiros a la cabeza (Headshot %) y el daño promedio por ronda (ADR) de varios jugadores en una partida de Valorant. Elegí el gráfico de dispersión porque es perfecto para descubrir si existe una relación entre dos variables.

```python
headshot_pct = [15, 18, 22, 25, 28, 30, 35, 40]
dano_promedio = [110, 125, 130, 145, 150, 165, 180, 195]

plt.figure(figsize=(8, 5))

plt.scatter(headshot_pct, dano_promedio, c='purple', s=100, alpha=0.7, marker='^', label='Jugadores')

plt.title('Relación entre Precisión y Daño en partida')
plt.xlabel('Headshot (%)')
plt.ylabel('Daño Promedio por Ronda (ADR)')
plt.legend()
plt.grid(True)

plt.savefig('imgs/valorant.png', dpi=300, bbox_inches='tight')
plt.show()
```

### 4. Histograma: Distribución de Notas de Programación
Este histograma muestra cómo se distribuyen las notas del primer parcial de programación en un curso de 50 alumnos. Elegí el histograma porque me pareció genial para agrupar datos continuos en rangos y ver dónde se concentra la nota mas común por decirlo de alguna manera.

```python
np.random.seed(42) 
notas_examen = np.random.normal(loc=7.5, scale=1.5, size=50) 
notas_examen = np.clip(notas_examen, 0, 10)

plt.figure(figsize=(8, 5))

plt.hist(notas_examen, bins=8, color='green', edgecolor='black', alpha=0.6, label='Alumnos')

plt.title('Distribución de Notas del Parcial')
plt.xlabel('Calificación (0 a 10)')
plt.ylabel('Cantidad de Alumnos')
plt.legend()

plt.savefig('imgs/notas.png', dpi=300, bbox_inches='tight')
plt.show()
```

### 5. Gráfico de Torta: Distribución del tiempo de estudio
Este gráfico visualiza cómo se reparte el tiempo total de estudio semanal entre distintas materias que estoy cursando en la Universidad Austral. Elegí el gráfico de torta porque ilustra de forma muy clara cómo un todo (el 100% de las horas) se divide en distintas proporciones o porcentajes.

```python
materias = ['Programación', 'Álgebra', 'Introducción a la Ingeniería', 'Otras']
horas = [12, 8, 6, 4]
colores_pie = ['#ff9999','#66b3ff','#99ff99','#ffcc99']
separacion = (0.1, 0, 0, 0) #

plt.figure(figsize=(7, 7))

plt.pie(horas, labels=materias, colors=colores_pie, explode=separacion, autopct='%1.1f%%', shadow=True, startangle=90)

plt.title('Distribución Semanal de Horas de Estudio')

plt.savefig('imgs/materias.png', dpi=300, bbox_inches='tight')
plt.show()
```
